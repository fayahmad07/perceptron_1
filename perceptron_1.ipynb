```python
def linear_sparable(xrange,yrange,coeffs):
  xmin = xrange[0]; xmax = xrange[1]; dx = xrange[2]
  ymin = yrange[0]; ymax = yrange[1]; dy = yrange[2]
  a = coeffs[0]; b = coeffs[1]; c = coeffs[2]
  xx = []; yy= []; zz = []
  for y in range (ymin,ymax,dy):
    for x in range (xmin, xmax, dx):
      xx.append(x); yy.append(y)
      if a*x+b*y+c>=0:
        zz.append(1)
      else:
        zz.append(0)
  return (xx,yy,zz)
```

```python
def plot_dataset(plt,x,y,z,titel):
  styles = ['s','o']
  fill = ['#ddf','#fdd']; stroke = ['b','r']
  plt.figure(figsize=(4,4))

  for i in range (len(x)):
    plt.plot(x[i], y[i], marker=styles[z[i]], mec= fill[z[i]], mfc =
             stroke[z[i]])
  plt.xticks([i*5 for i in range (5)]); plt.xlabel('$x_1$')
  plt.yticks([i*5 for i in range (5)]); plt.ylabel('$x_2$')
  plt.title(f'{titel}');
  plt.grid();
  plt.show()
```

```python
import matplotlib.pyplot as plt
x, y,z = linear_sparable([0,20,1],[0,20,1],[1,0,-10])
plot_dataset(plt, x,y,z,f'Linear equation: $x-10=0$')
```

```python
def binary_step(x):
  if x>=0:
    return 1
  else:
    return 0
def perseptron(w,x):
  ww=np.array(w)
  xx=np.array(x+[1])
  return binary_step(np.dot(ww,xx))
```

# **Decision Boundary Line**

```python
# @title Default title text
def plot_dataset__(plt,x,y,z,titel,xb,yb):
  styles = ['s','o']
  fill = ['#ddf','#fdd']; stroke = ['b','r']
  plt.figure(figsize=(4,4))

  for i in range (len(x)):
    plt.plot(x[i], y[i], marker=styles[z[i]], mec= fill[z[i]], mfc =
             stroke[z[i]])
  plt.plot(xb,yb, 'g--',lw=2)
  plt.xticks([i*5 for i in range (5)]); plt.xlabel('$x_1$')
  plt.yticks([i*5 for i in range (3)]); plt.ylabel('$x_2$')
  plt.title(f'{titel}');
  plt.grid();
  plt.show()
```

```python
weights = [1, -2, -1]
xx=[]; yy=[]; zz=[]
for x2 in range (11):
  for x1 in range (20):
    x = [x1, x2]
    y = perseptron(weights,x)
    xx.append(x1); yy.append(x2); zz.append(y)
plot_dataset_(plt,xx,yy,zz,'$w=[1, -2, -1]$')
```

```python
def decision_boundary(xr,yr,w):
  a = w[0]; b=w[1]; c=w[2]
  xmin= xr[0]; xmax=xr[1]
  ymin= yr[0]; ymax=yr[1]

  xx=[]; yy=[]
  for x in range(xmin, xmax+1):
    for y in range(xmin,xmax+1):
      z=a*x+b*y+c
      if z==0:
        xx.append(x)
        yy.append(y)
  return (xx,yy)
```

```python
weights = [1, -2, -1]
xx=[]; yy=[]; zz=[]
for x2 in range (11):
  for x1 in range (20):
    x = [x1, x2]
    y = perseptron(weights,x)
    xx.append(x1); yy.append(x2); zz.append(y)

xr=[0,20]; yr=[0,10]
xb, yb = decision_boundary(xr,yr,weights)
plot_dataset__(plt,xx,yy,zz,'$w=[1, -2, -1]$',xb,yb)
plt.show()
```

```python
weights = [1, -2, -1]

xr=[0,20]; yr=[0,10]
xb, yb = decision_boundary(xr,yr,weights)
plt.plot(xb,yb, 'g--', lw=2)
plt.xlim([0,20]); plt.ylim([0,10])
plt.xlabel(['$x_1$']); plt.ylabel('$x_2$')
plt.show()
```
